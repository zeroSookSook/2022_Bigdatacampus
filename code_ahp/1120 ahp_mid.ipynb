```python
import pandas as pd
import numpy as np

from sklearn import preprocessing

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
#등급 계산할 함수
def minus(m):
    if m>7/8:
        rank=1/9
    elif 6/8 <m<= 7/8:
        rank=1/8
    elif 5/8 <m<= 6/8:
        rank=1/7
    elif 4/8 <m<= 5/8:
        rank=1/6
    elif 3/8 <m<= 4/8:
        rank=1/5
    elif 2/8 <m<= 3/8:
        rank=1/4
    elif 1/8 <m<= 2/8:
        rank=1/3
    elif 0 <m<= 1/8:
        rank=1/2
    elif m==0:
        rank=1
    elif -1/8 <=m< 0:
        rank=2
    elif -2/8 <=m< -1/8:
        rank=3
    elif -3/8 <=m< -2/8:
        rank=4
    elif -4/8 <=m< -3/8:
        rank=5
    elif -5/8 <=m< -4/8:
        rank=6
    elif -6/8 <=m< -5/8:
        rank=7
    elif -7/8 <=m< -6/8:
        rank=8
    elif m< -7/8:
        rank=9
    else:
        pass
    return rank
```
