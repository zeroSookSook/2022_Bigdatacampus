```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
four=pd.read_excel('ahp_four.xlsx')
four=four.rename(columns={'Unnamed: 0':'dong_nm'})
four.set_index("dong_nm",inplace=True)

sum_rank=pd.read_excel('ahp_sum_rank.xlsx')
sum_rank=sum_rank.rename(columns={'Unnamed: 0':'dong_nm'})
sum_rank.set_index("dong_nm",inplace=True)

one_ratio=pd.read_excel('ahp_one_ratio.xlsx')
one_ratio=one_ratio.rename(columns={'Unnamed: 0':'dong_nm'})
one_ratio.set_index("dong_nm",inplace=True)

cc_value=pd.read_excel('ahp_cc_value.xlsx')
cc_value=cc_value.rename(columns={'Unnamed: 0':'dong_nm'})
cc_value.set_index("dong_nm",inplace=True)

sum_trans=pd.read_excel('ahp_sum_trans.xlsx')
sum_trans=sum_trans.rename(columns={'Unnamed: 0':'dong_nm'})
sum_trans.set_index("dong_nm",inplace=True)
```

#  four

```python
#열별 합 구하기
four=pd.concat([four,(four.sum(axis='rows')).to_frame().T])
four.rename(index={0:'계'},inplace=True)
four
```

```python
#열별 합 1로 맞추고 행별 평균 구하기
four_norm=four.iloc[:5,:] / four.iloc[4]
four_norm

f_four=pd.concat([four_norm,pd.Series(four_norm.mean(axis='columns')).to_frame()],axis=1)
f_four.rename(columns={0:'상대비중'},inplace=True)
f_four
```

# sum_rank

```python
#열별 합 구하기
sum_rank=pd.concat([sum_rank,(sum_rank.sum(axis='rows')).to_frame().T])
sum_rank.rename(index={0:'계'},inplace=True)
```

```python
#열별 합 1로 맞추고 행별 평균 구하기
sum_rank_norm=sum_rank.iloc[:66,:] / sum_rank.iloc[65]
sum_rank_norm

f_sum_rank=pd.concat([sum_rank_norm,pd.Series(sum_rank_norm.mean(axis='columns')).to_frame()],axis=1)
f_sum_rank.rename(columns={0:'환경_상대비중'},inplace=True)
f_sum_rank
```

# one_ratio

```python
#열별 합 구하기
one_ratio=pd.concat([one_ratio,(one_ratio.sum(axis='rows')).to_frame().T])
one_ratio.rename(index={0:'계'},inplace=True)
```

```python
#열별 합 1로 맞추고 행별 평균 구하기
one_ratio_norm=one_ratio.iloc[:66,:] / one_ratio.iloc[65]
one_ratio_norm

f_one_ratio=pd.concat([one_ratio_norm,pd.Series(one_ratio_norm.mean(axis='columns')).to_frame()],axis=1)
f_one_ratio.rename(columns={0:'1인가구_상대비중'},inplace=True)
f_one_ratio
```

# cc_value

```python
#열별 합 구하기
cc_value=pd.concat([cc_value,(cc_value.sum(axis='rows')).to_frame().T])
cc_value.rename(index={0:'계'},inplace=True)
```

```python
#열별 합 1로 맞추고 행별 평균 구하기
cc_value_norm=cc_value.iloc[:66,:] / cc_value.iloc[65]
cc_value_norm

f_cc_value=pd.concat([cc_value_norm,pd.Series(cc_value_norm.mean(axis='columns')).to_frame()],axis=1)
f_cc_value.rename(columns={0:'상권_상대비중'},inplace=True)
f_cc_value
```

#  sum_trans

```python
#열별 합 구하기
sum_trans=pd.concat([sum_trans,(sum_trans.sum(axis='rows')).to_frame().T])
sum_trans.rename(index={0:'계'},inplace=True)
```

```python
#열별 합 1로 맞추고 행별 평균 구하기
sum_trans_norm=sum_trans.iloc[:66,:] / sum_trans.iloc[65]
sum_trans_norm

f_sum_trans=pd.concat([sum_trans_norm,pd.Series(sum_trans_norm.mean(axis='columns')).to_frame()],axis=1)
f_sum_trans.rename(columns={0:'유동인구_상대비중'},inplace=True)
f_sum_trans
```

# 최적 상대적 중요도 도출

```python
# 요인별 상대비중

sdbj=f_four[['상대비중']].iloc[:4,:]
sdbj
```

```python
# 행정동별 상대비중

dsdbj=pd.concat([f_sum_rank[['환경_상대비중']],f_one_ratio[['1인가구_상대비중']],\
           f_cc_value[['상권_상대비중']],f_sum_trans[['유동인구_상대비중']]],axis=1)
dsdbj=dsdbj.iloc[:65,:]
dsdbj
```

```python
# 두 행렬 곱해서 최종점수 도출

ff=pd.DataFrame(np.dot(dsdbj,sdbj))
ff.index=['개포1동', '개포2동', '개포4동', '논현1동', '논현2동', '대치1동', '대치2동', '대치4동', '도곡1동',
       '도곡2동', '삼성1동', '삼성2동', '세곡동', '수서동', '신사동', '압구정동', '역삼1동', '역삼2동',
       '일원1동', '일원2동', '일원본동', '청담동', '가양1동', '가양2동', '가양3동', '공항동', '등촌1동',
       '등촌2동', '등촌3동', '발산1동', '방화1동', '방화2동', '방화3동', '염창동', '우장산동', '화곡1동',
       '화곡2동', '화곡3동', '화곡4동', '화곡8동', '화곡본동', '가락1동', '가락2동',
       '거여1동', '거여2동', '마천1동', '마천2동', '문정1동', '문정2동', '방이1동', '방이2동', '삼전동',
       '송파1동', '송파2동', '오금동', '오륜동', '위례동', '잠실2동', '잠실3동', '잠실4동',
       '잠실6동', '잠실7동', '잠실본동', '장지동', '풍납1동']
ff.rename(columns={0:'최종점수'},inplace=True)

fff=pd.concat([dsdbj,ff],axis=1)
fff=fff.sort_values(['최종점수'],ascending=False)
fff
```
