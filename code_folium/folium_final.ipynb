```python
import requests
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import os
import webbrowser
import folium
from folium import plugins
```

```python
#시각화할 지도

#자치구
state_geo = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
#행정동
state_geo2='https://raw.githubusercontent.com/vuski/admdongkor/master/ver20220309/HangJeongDong_ver20220309.geojson'
```

```python
#자치구
df_goo=pd.read_excel('fin_dataset_goo.xlsx')
df_goo
```

```python
#행정동
df_dong=pd.read_excel('fin_dataset_dong.xlsx')
df_dong
```

# 자치구별

```python
# 전체가구수

m = folium.Map(location=[37.562225, 126.978555], tiles="OpenStreetMap", zoom_start=11)
m.choropleth(
    geo_data=state_geo,
    name='전체가구',
    data=df_goo,
    columns=['goo_nm', 'sum_hh'],
    key_on='feature.properties.name',
    fill_color='Blues',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = 'sum_hh'
)

m.save('kr_incode.html')
m
```
